```python
# Matplotlib
# Matplotlibとは
"""
Pythonのグラフ描写のライブラリ
"""
```

```python
# Matplotlibの使い方
"""
matplotlib.pyplotをインポートする
x軸の配列を作る
y軸の配列を作る
plot関数を使いplotする
show関数を使いplotしたグラフを描写する
"""
import numpy as np
import matplotlib.pyplot as plt

# -5から5まで、0.1区切りの配列を作る
x = np.arange(-5, 5, 0.1)
# 配列xに対して、それぞれsin(x)を求めてy軸の配列を生成
y = np.sin(x)

plt.plot(x, y)
plt.show()
```
