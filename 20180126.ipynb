```python
# Matplotlib
# Matplotlibとは
"""
Pythonのグラフ描写のライブラリ
"""
```

```python
# Matplotlibの使い方
"""
matplotlib.pyplotをインポートする
x軸の配列を作る
y軸の配列を作る
plot関数を使いplotする
show関数を使いplotしたグラフを描写する
"""
import numpy as np
import matplotlib.pyplot as plt

# -5から5まで、0.1区切りの配列を作る
x = np.arange(-5, 5, 0.1)
# 配列xに対して、それぞれsin(x)を求めてy軸の配列を生成
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# 色々なグラフの描画 / 折れ線グラフ
import matplotlib.pyplot as plt

data = [2, 4, -1, 3, 7, 4, 9]
plt.plot(data)
plt.show()
```

```python
# plot()
"""
第一引数にx軸
第二引数にy軸
第三引数のオプション。デフォルトは各点が線で繋がっている
"o" - 点
"ro" - 赤い点
"go" - 緑の点
"b--" - 青い点線
"^" - 三角の点
"s" - 四角の点
"""
import matplotlib.pyplot as plt
x_data = [20, 40, 60, 80]
y_data = [10, 5, 3, 1]
plt.plot(x_data, y_data, marker='^')
plt.show()
```
