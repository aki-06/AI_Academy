```python
# 欠損値の補完
"""
fillna()
    - 引数として与えた数をNaNに代入
fillna(method='ffill')
    - 前の値で埋める
"""
import numpy as np
from numpy import nan as NA
import pandas as pd

np.random.seed(0)

df = pd.DataFrame(np.random.rand(10, 4))
df.iloc[1, 0] = NA
df.iloc[6:, 2] = NA

df.fillna(method='ffill')
```

```python
# 欠損値の補完（平均値代入法）
"""
平均値代入法
    - 欠損値をその列や行を平均値で埋める
"""
import pandas as pd
from numpy import nan as NA
import numpy as np

np.random.seed(0)

df = pd.DataFrame(np.random.rand(10, 4))

df.iloc[1, 0] = NA
df.iloc[2, 2] = NA
df.iloc[5:, 3] = NA
df
```

```python
df.fillna(df.mean())
```
