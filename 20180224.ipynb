```python
# 3D Axesを作成する
"""
from mpl_toolkits.mplot3d import Axes3D
projection='3d'
"""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(X, Y, Z)
plt.show()
```

```python
# 曲面を作成する
"""
ax.plot_surface()
"""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X ** 2 + Y ** 2)/2) / (2*np.pi)

# figureオブジェクト作成
fig = plt.figure(figsize=(6, 6))

# サブプロット作成
ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(X, Y, Z)
plt.show()
```

```python
# 3Dヒストグラムを作成する
"""
bar3d(xpos, ypos, zpos, dx, dy, dz)
"""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111, projection='3d')

xpos = [i for i in range(10)]
ypos = [i for i in range(10)]
zpos = np.zeros(10)

dx = np.ones(10)
dy = np.ones(10)
dz = [i for i in range(10)]

ax1.bar3d(xpos, ypos, zpos, dx, dy, dz)
plt.show()
```

```python
# 3D散布図を作成する
"""
scatter3D()
np.ravel
    - データを一次元配列にする
"""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)

X = np.random.randn(1000)
Y = np.random.randn(1000)
Z = np.random.randn(1000)

fig = plt.figure(figsize=(5, 5))

ax = fig.add_subplot(1, 1, 1, projection='3d')

x = np.ravel(X)
y = np.ravel(Y)
z = np.ravel(Z)

ax.scatter3D(x, y, z)

plt.show()
```

```python
# 3Dグラフにカラーマップを適用する
"""
from matplotlib import cm
cmap = cm.coolwarm
    - zの値についてカラーマップを適用する
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()
```

```python
# ホールドアウト法
from sklearn import datasets
from sklearn.model_selection import train_test_split

# irisというデータセットの読み込み
iris = datasets.load_iris()
X = iris.data
y = iris.target

# データを格納
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print('X_train: ', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train: ', y_train.shape)
print('y_train: ', y_train.shape)
```

```python
# クロスバリデーション
import numpy as np
from sklearn import svm, datasets, cross_validation

iris = datasets.load_iris()
X = iris.data
y = iris.target

# 機械学習アルゴリズムSVMを使用
svc = svm.SVC(C=1, kernel='rbf', gamma=0.001)

scores = cross_validation.cross_val_score(svc, X, y, cv=5)

print(scores)
print('平均: ', scores.mean())
```
