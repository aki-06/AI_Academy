```python
import pandas as pd

# csv読み込み
df = pd.read_csv('20180307_sample.csv', encoding='cp932')
df
```

```python
def merge_answer(ds_index):
    """
    回答(修正)がNaNの場合、回答を回答(修正)に代入する
    """
    ds = ds_index.copy()
    if ds['回答(修正)'] != ds['回答(修正)']:
        ds['回答(修正)'] = ds['回答']
        
    return ds

df = df.apply(merge_answer, axis=1)
df
```

```python
# 「回答」は不要なのでdrop
fixed_df = df.drop(['回答'], axis=1)
fixed_df
```
