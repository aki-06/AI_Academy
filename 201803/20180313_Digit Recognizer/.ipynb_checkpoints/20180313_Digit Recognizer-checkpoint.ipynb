```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier


# データの読み込み
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
# データ数の確認
train.shape
```

```python
train_null = train.isnull().sum()
train_null

train_null_sum = train_null.sum()
train_null_sum # 欠損値なし
```

```python
y_train = train['label']
X_train = train.drop(labels=['label'], axis=1)
X_train.head()
```

```python
X_train_img = X_train.iloc[1].as_matrix()
X_train_img = X_train_img.reshape((28, 28))
plt.imshow(X_train_img, cmap='Greys')
plt.axis('off')
plt.title('label_number:' + str(y_train.iloc[1]))
plt.show()
```

```python
rfc = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)
X_test = test
predicted_numbers = rfc.predict(X_train)
predicted_numbers
```

```python
i = 2
X_test_img = X_test.iloc[i].as_matrix()
X_test_img = X_test_img.reshape((28, 28))
plt.imshow(X_test_img, cmap='Greys')
plt.axis('off')
plt.title('predicted_number:' + str(predicted_numbers[i]))
plt.show()
```

```python
result_data = pd.DataFrame({'ImageId': range(1, len(predicted_numbers)+1), 'Label': predicted_numbers})
result_data = result_data.iloc[0:28000]
# result_data.head()
# result_data.tail()
result_data.to_csv('submission.csv', sep=',', index=False)
```
