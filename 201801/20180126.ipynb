```python
# Matplotlib
# Matplotlibとは
"""
Pythonのグラフ描写のライブラリ
"""
```

```python
# Matplotlibの使い方
"""
matplotlib.pyplotをインポートする
x軸の配列を作る
y軸の配列を作る
plot関数を使いplotする
show関数を使いplotしたグラフを描写する
"""
import numpy as np
import matplotlib.pyplot as plt

# -5から5まで、0.1区切りの配列を作る
x = np.arange(-5, 5, 0.1)
# 配列xに対して、それぞれsin(x)を求めてy軸の配列を生成
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# 色々なグラフの描画 / 折れ線グラフ
import matplotlib.pyplot as plt

data = [2, 4, -1, 3, 7, 4, 9]
plt.plot(data)
plt.show()
```

```python
# plot()
"""
第一引数にx軸
第二引数にy軸
第三引数のオプション。デフォルトは各点が線で繋がっている
"o" - 点
"ro" - 赤い点
"go" - 緑の点
"b--" - 青い点線
"^" - 三角の点
"s" - 四角の点
"""
import matplotlib.pyplot as plt
x_data = [20, 40, 60, 80]
y_data = [10, 5, 3, 1]
plt.plot(x_data, y_data, marker='^')
plt.show()
```

```python
# 棒グラフ
"""
bar()
"""
import matplotlib.pyplot as plt
x = range(0, 7)
y = [20, 40, 17, 99, 50, 39, 74]
plt.bar(x, y)
plt.show()
```

```python
# 散布図
"""
scatter()
"""
import matplotlib.pyplot as plt

x = [2, 6, 8]
y = [4, 9, 1]
plt.scatter(x, y)
plt.show()
```

```python
# 円グラフ
"""
pie()
"""
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
data = [12, 43, 89, 20, 44]
ex = [0, 1, 0, 0, 0]
plt.pie(data, explode=ex, labels=labels, autopct='%1.1f%%', counterclock=False)
plt.show()
```

```python
# 複数のグラフを描画する
"""
add_subplot()
figure()
"""
import matplotlib.pyplot as plt

x1, y1 = range(0, 5), [10, 41, 60, 33, 19]
x2, y2 = range(0, 5), [2, 57, 41, 49, 10]

fig = plt.figure()

a1 = fig.add_subplot(1, 2, 1) # 第3引数の1は左側を意味しています
a1.bar(x1, y1)
a1.set_title('A')

a2 = fig.add_subplot(1, 2, 2) # 第3引数の2は右側を意味しています
a2.bar(x2, y2)
a2.set_title('B')

plt.show()
```
