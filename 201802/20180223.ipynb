```python
# 値に応じてマーカーの大きさを設定する
"""
matplotlib.pyplot.scatter()
    - x, yと順番に対応するリスト型のデータをs=データで受け渡す
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

plt.scatter(x, y, s=z)
plt.show()
```

```python
# 値の応じてマーカーの濃さを設定する
"""
c=濃さに対応するデータ(リストなど)
cmap='色系統'
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

plt.scatter(x, y, c=z, cmap='Blues')
plt.show()
```

```python
# カラーバーを表示する
"""
matplotlib.pyplot.colorbar()
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

plt.scatter(x, y, c=z, cmap='Reds')

plt.colorbar()

plt.show()
```

```python
# 円グラフを作成する
"""
matpltlib.pyplot.pie()
matplotlib.pyplot.axis('equal')
    - 円形
"""
import matplotlib.pyplot as plt

data = [10, 30, 2, 41, 19]

plt.pie(data)
plt.axis('equal')
plt.show()
```

```python
# 円グラフにラベルを設定する
"""
matplotlib.pyplot.pie(データ, labels=ラベル)
"""
import matplotlib.pyplot as plt

data = [20, 30, 15, 22, 8]
labels= ['apple', 'banana', 'grape', 'orange', 'mango']

plt.pie(data, labels=labels)
plt.axis('equal')
plt.show()
```

```python
# 特定の要素を目立たせる
"""
目立たせたい度合い(0から1の値)を目立たせたい要素のリストと同じ順番を
explodeとして渡す
"""
import matplotlib.pyplot as plt

data = [21, 13, 10, 9, 30]
labels = ['apple', 'banana', 'grape', 'orange', 'mango']
explode = [0, 0, 0.1, 0, 0]

plt.pie(data, labels=labels, explode=explode)
plt.axis('equal')
plt.show()
```
