```python
# 棒グラフを作成する
"""
matplotlib.pyplot.bar()
"""

import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [12, 45, 32, 6, 22, 18]

plt.bar(x, y)
plt.show()
```

```python
# 横軸にラベルを設定する
"""
tick_label
"""
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [12, 32, 56, 9, 38, 44]

labels = ['apple', 'banana', 'peach', 'grape', 'mango', 'orange']

plt.bar(x, y, tick_label=labels)
plt.show()
```

```python
# 積み上げ棒グラフを作成する
"""
matplotlib.pyplot.bar()のbottomにデータ列を渡す
plt.legend(('y1', 'y2'))
    - 系統ラベルを指定
"""
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]

plt.bar(x, y1, tick_label=labels)
plt.bar(x, y2, bottom=y1)

plt.legend(('y1', 'y2'))
plt.show()
```

```python
# ヒストグラムを作成する
"""
matplotlib.pyplot.hist()
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.randn(10000)

plt.hist(data)
plt.show()
```

```python
# ビン数を設定する
"""
binsを指定する
    - 任意の数のサイズに等しい階級に分ける
bins='auto'
    - ビン数を自動に設定
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.randn(10000)

plt.hist(data, bins=100)
plt.show()
```

```python
# 正規化を行う
"""
正規化
    - ヒストグラムの分布を正規分布と仮定した時、合計値が１となるようにヒストグラムを操作すること
normed=True
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.randn(10000)

plt.hist(data, bins=100, normed=True)
plt.show()
```

```python
# 累積ヒストグラムを作成する
"""
累積ヒストグラム
    - 最低階級から、階級が上がるごとの数を積み上げていくヒストグラム
cumulative=True
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.randn(10000)

plt.hist(data, bins=100, normed=True, cumulative=True)
plt.show()
```
