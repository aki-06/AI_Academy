```python
# 一部の行を得る
"""
df.head()
    - 冒頭の５行を返す
df.tail()
    - 末尾の５行を返す
"""
import pandas as pd
import numpy as np

np.random.seed(0)
columns = ['apple', 'banana', 'peach', 'mango', 'grape']

df = pd.DataFrame()
for column in columns:
    df[column] = np.random.choice(range(1, 11), 10)
df.index = range(1, 11)

df
```

```python
# 末尾５行出力
df.tail()
```

```python
# 計算処理を適用する
import pandas as pd
import numpy as np
import math

np.random.seed(0)
columns = ['apple', 'banana', 'grape', 'mango', 'peach']

df = pd.DataFrame()
for column in columns:
    df[column] = np.random.choice(range(1, 11), 10)
df.index = range(1, 11)

df
```

```python
# 値が２倍
double_df = df * 2
double_df.head()
```

```python
# 値が２乗
square_df = df ** 2
square_df.head()
```

```python
# 平方根
sqrt_df = np.sqrt(df)
sqrt_df.head()
```

```python
# 要約統計量を得る
"""
要約統計量
    - 列ごとの平均値、最大値、最小値などの統計情報をまとめたもの
df.describe()
    - 列ごとの個数、平均値、標準偏差、最小値、四分位数、最大値を含むDataFrameを返す
"""
import pandas as pd
import numpy as np

np.random.seed(0)
columns = ['apple', 'banana', 'grape', 'peach', 'mango']

for column in columns:
    df[column] = np.random.choice(range(1, 11), 10)
df.index = range(1, 11)

df.head()
```

```python
df_mean = df.describe().loc[['mean', 'max', 'min']]
df_mean
```

```python
# DataFrameの行間または列間の差を求める
"""
df.diff('行または列の間隔', axis='方向')
第一引数が正
    - 前の行との差
第一引数が負
    - 後の行との差
axis=0
    - 行方向
axis=1
    - 列方向
"""
import pandas as pd
import numpy as np

columns = ['apple', 'banana', 'grape', 'peach', 'mango']
df = pd.DataFrame()
for column in columns:
    df[column] = np.random.choice(range(1, 11), 10)
df.index = range(1, 11)

df.head()
```

```python
df_diff = df.diff('-2', axis=0)
df_diff
```

```python
df_diff = df.diff('-1', axis=1)
df_diff
```

```python
# グループ化
"""
df.groupby()
"""
import pandas as pd

# 一部の都道府県に関するDataFrameを作成
prefecture_df = pd.DataFrame([["Tokyo", 2190, 13636, "Kanto"], ["Kanagawa", 2415, 9145, "Kanto"],
                              ["Osaka", 1904, 8837, "Kinki"], ["Kyoto", 4610, 2605, "Kinki"],
                              ["Aichi", 5172, 7505, "Chubu"]], 
                             columns=["Prefecture", "Area", "Population", "Region"])
# 出力
prefecture_df
```

```python
grouped_df = prefecture_df.groupby('Region')
mean_df = grouped_df.mean()
mean_df
```
