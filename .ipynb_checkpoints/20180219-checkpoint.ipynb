```python
# サブプロット内のグラフの表示範囲を設定する
"""
グラフの表示範囲を設定する
    - set_xlim(範囲)
    - set_ylim(範囲)
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi)
y = np.sin(x)

fig = plt.figure(figsize=(6, 9))
ax = fig.add_subplot(2, 3, 5)
plt.subplots_adjust(wspace=1, hspace=1)
ax.set_ylim([0, 1])

ax.plot(x, y)

axi=[]
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i+1)

plt.show()
```
