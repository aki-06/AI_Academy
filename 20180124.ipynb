```python
# random
"""
乱数を生成する。
numpy.random.rand()で0〜1の一様乱数を生成する。
引数を指定すれば、複数の乱数を生成できる。
"""
import numpy as np

print(np.random.rand()) # 0から1の乱数を一つ生成

token = np.random.rand(100)
print(token) # 0〜1の乱数を100個生成

token2 = np.random.rand(10, 10)
print(token2) # 0〜1の乱数を10*10の行列を生成

token3 = np.random.rand(100) * 40 + 70
print(token3)
```

```python
import numpy as np

"""
標準正規分布
ガウシアン。
標準正規分布はrandn()
平均・分散を指定したい場合はnormal()
"""
np.random.randn() # 標準正規分布 (平均0, 標準偏差1)
np.random.randn(10) # 標準正規分布を10個生成
np.random.randn(10, 10) # 標準正規分布による 10x10 の行列
np.random.normal(50, 10) # 平均50、標準偏差10の正規分布

# 二項分布
np.random.binomial(n=100, p=0.5)

# ボアソン分布
"""
稀にしか起きない現象を長時間観測したときに起きる回数の分布。
λはその平均。
"""
np.random.poisson(lam=10)

# ベータ分布
"""
二項分布の共役事前分布などに用いる。
"""
np.random.beta(a=3, b=5)
```

```python
import numpy as np

np.random.randint(100) # 0〜99 の整数を1個生成
np.random.randint(30, 70) # 30〜69 の整数を1個生成
np.random.randint(0, 100, 20) # 0〜99 の整数を20個生成
np.random.randint(0, 100, (5, 5)) # 0〜99 の整数で5x5の行列を生成

np.random.random_integers(100) # 1〜100 の整数を1個生成
np.random.random_integers(30, 70) # 30〜70 の整数を1個生成

"""
Numpyで生成される乱数は擬似乱数であるため、
固定したseedを指定すれば毎回同じ乱数がでる。
デバック時など、毎回同じ乱数を出して欲しいときに使う。
"""

np.random.seed(0)
np.random.rand()
```
