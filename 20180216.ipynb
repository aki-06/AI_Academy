```python
# 一部の行を得る
"""
df.head()
    - 冒頭の５行を返す
df.tail()
    - 末尾の５行を返す
"""
import pandas as pd
import numpy as np

np.random.seed(0)
columns = ['apple', 'banana', 'peach', 'mango', 'grape']

df = pd.DataFrame()
for column in columns:
    df[column] = np.random.choice(range(1, 11), 10)
df.index = range(1, 11)

df
```

```python
# 先頭５行出力
df.head()
```

```python
# 先頭３行出力
df.head(3)
```

```python
# 末尾５行出力
df.tail()
```
